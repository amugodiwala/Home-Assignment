```python
#Loading the Numeric Values file and separating Positive Vexel Values
import nibabel as nb

numeric_file= "attention_uniformity-test_z_FDR_0.01.nii"
img=nb.load(numeric_file)
data= img.get_fdata() 
poss_value= data[data>0] 
new_numeric_file= nb.Nifti1Image(data, img.affine)

#Just to see if it works
print(type(new_numeric_file))
```
